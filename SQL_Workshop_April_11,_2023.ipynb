#Introduction

1) Structured - if you can organise data into rows and columns, it is structured.
2) Semi-structured - if you cannot organise data into rows and columns, it is semi-structured.
3) Unstructured - if you cannot organise data (and it is super difficult) into rows and columns, it is unstructured.

#SQL Command Types

1) DQL (Data Query Language)
-> SELECT
2) DDL (D Definition L)
-> CREATE, DROP, ALTER, RENAME
3) DML (D Manipulation L)
-> INSERT, UPDATE, DELETE
4) TCL (Transaction Control L)
-> COMMIT, ROLLBACK, SAVEPOINT
5) DCL D Control L)
-> GRANT, REVOKE

Note: TCL and DCL are only important for Data Engineers (not for Data Scientists).
